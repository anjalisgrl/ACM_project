```
#LexLuthor_1806285_SnakesAndLaddersAssignment
import random
import sys

MAX_VAL = 100

snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79}

ladders = {3 : 38, 24 : 33, 42 : 93, 72 : 84}


def get_player_names():
    print("###### Welcome to Snakes & Ladders Game #####")
    player1_name1 = None
    while not player1_name1:
        player1_name1 = input("Enter the name of Player 1: ").strip()

    player2_name1 = None
    while not player2_name1:
        player2_name1 = input("Enter the name of Player 2: ").strip()

    print("###### Let us start #####")
    player1_name= "Player 1"
    player2_name = "Player 2"
    return player1_name, player2_name


def got_snake_bite(old_value, current_value):
    print(" Your final position is " + str(current_value))


def got_ladder_jump(old_value, current_value):
    print(" Your final position is " + str(current_value))

def get_dice_value1(player_name):
    
  d=input(player_name +" : ")
  if d=="roll":
      dice_value = random.randint(1, 6)
      print("You got a " + str(dice_value))
      #dice_value = r + dice_value
      return dice_value
  
  elif d=="quit":
    print("Player 2 won the game")
    print("###### Game Successfully Finished #####")
    sys.exit(1)
 
  else:
    dice_value = int(d)
    print("You got a ",dice_value)
    while dice_value<1 or dice_value>20:
      print(dice_value)
    return dice_value

def get_dice_value2(player_name):
    
  d=input(player_name +" : ")
  if d=="roll":
      dice_value = random.randint(1, 6)
      print("You got a " + str(dice_value))
      return dice_value
  
  elif d=="quit":
    print("Player 1 won the game")
    print("###### Game Successfully Finished #####")
    sys.exit(1)
 
  else:
    dice_value = int(d)
    print("You got a ",dice_value)
    while dice_value<1 or dice_value>20:
      print(dice_value)
    return dice_value


def snake_ladder(player_name, current_value, dice_value):
    old_value = current_value
    current_value = current_value + dice_value

    if current_value > MAX_VAL:
        print(" Your final position is " + str(old_value))
        return old_value

    if current_value in snakes:
        final_value = snakes.get(current_value)
        got_snake_bite(current_value, final_value)

    elif current_value in ladders:
        final_value = ladders.get(current_value)
        got_ladder_jump(current_value, final_value)

    else:
        final_value = current_value
        print(" Your final position is " + str(current_value))
    return final_value


def check_win(player_name, position):
    if MAX_VAL == position:
        print(player_name+ " won the game")
        print("###### Game Successfully Finished #####")
        sys.exit(1)


def start():
    
    player1_name, player2_name = get_player_names()
   

    player1_current_position = 0
    player2_current_position = 0

    while True:
        
      
        dice_value = get_dice_value1(player1_name)
        player1_current_position = snake_ladder(player1_name, player1_current_position, dice_value)
        check_win(player1_name, player1_current_position)

        dice_value = get_dice_value2(player2_name)
        player2_current_position = snake_ladder(player2_name, player2_current_position, dice_value)
        check_win(player2_name, player2_current_position)


if __name__ == "__main__":
    start()
```
